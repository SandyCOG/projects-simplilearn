```python
# Drop irrelevant columns
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
# Renaming the column names
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
#check for duplicates
duplicate_rows = df[df.duplicated()]
print('Number of duplicated rows: ', duplicate_rows.shape)
```

```python
#count the rows
df.count()
```

```python
df = df.drop_duplicates()
df.count()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#drop missing values
df = df.dropna()
df.count()
```

```python
df.isnull().sum()
```

```python
#finding outliers
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
#plotting a pie chart for the number of cars per brand
df.Make.value_counts().nlargest(30).plot(kind='pie',figsize=(10,15))
plt.title('Number of cars per Make')
plt.xlabel('Make')
plt.ylabel('Number of cars')
```

```python
#plotting a histogram for the number of cars per brand
df.Make.value_counts().nlargest(40).plot(kind='line',figsize=(10,15))
plt.title('Number of cars per Make')
plt.xlabel('Make')
plt.ylabel('Number of cars')
```

```python
#plotting a histogram for the number of cars per brand
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(15,10))
plt.title('Number of cars per Make')
plt.xlabel('Make')
plt.ylabel('Number of cars')
```

```python
#find the relationship between the variables
plt.figure(figsize=(20,10))
C = df.corr()
sns.heatmap(C, cmap='coolwarm',annot=True)
C
```
